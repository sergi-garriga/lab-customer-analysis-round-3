```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_csv("files_for_lab/csv_files/marketing_customer_analysis.csv")
df.round(2)
```

```python
# Show DataFrame info
df.info()
```

```python
# Describe DataFrame
df.describe(include="all").round(2)
```

```python
# Let's also generate a pairwise scatterplot to explore the data and detect some correlations between data.

sns.pairplot(data=df)
```

# Show a plot of the total number of responses.




```python
# We are going to develop a plot that allows us to compare the total responses No versus the Yes.

sns.countplot(data=df, x="Response")
```

# Show a plot of the response rate by the sales channel.

```python
sns.countplot(
    data=df, x="Sales Channel", hue= "Response")
```

# Show a plot of the response rate by the total claim amount.

```python
f, ax = plt.subplots(figsize=(15, 7.5))

sns.histplot(x="Total Claim Amount", bins=30,
                hue="Response", multiple="stack",
                data=df)
```

# Show a plot of the response rate by income.

```python
f, ax = plt.subplots(figsize=(15, 7.5))

sns.histplot(x="Income", bins=50,
                hue="Response", kde=True,
                data=df)
```

```python
# Response rate by Income and State.

plt.figure(figsize=(16,5))
sns.barplot(y="State", x="Income", hue='Response', data=df, order=["Arizona", "California", "Nevada", "Oregon", "Washington"])
plt.show()
```
